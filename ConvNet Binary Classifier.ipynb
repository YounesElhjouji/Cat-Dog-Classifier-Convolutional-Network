# Imoprts 

```python
import tensorflow as tf
import keras
import numpy as np
from os import walk
import PIL.Image as Im
from random import shuffle
import os
import pandas as pd
from tqdm import tqdm
```

# Hyperparameters

```python
epochs = 100
nodes_per_layer = [1024,512]
layers = []
batch_size = 64
dropout_rate = 0.3
num_classes = 10
```

# Preparing the data

```python
def imagePaths(folderName):
    f = []
    for root, dirs, files in walk(folderName):
        for filename in files:
            f.append(folderName+'/'+filename)
    shuffle(f)
    return f

# This function opens the images and returns an array of grayscale images
def openImages(paths):
    images = []
    print (len(paths))
    for path in tqdm(paths):
        image = Im.open(path).convert("RGB")
        images.append(np.array(image.resize((64,64))))
    return np.array(images)


# Get labels of the images from the paths
def getLabels(paths):
    labels = []
    for path in paths:
        if "cat" in path:
            labels.append(1)
        else:
            labels.append(0)
    return np.array(labels)


# Process train images by making them the same size and extracting their labels then turning them to
# 1D arrays and returning them
def processTrainImages(path):
    paths = imagePaths(path)
    labels = getLabels(paths)
    images = openImages(paths)
    return images,labels

def processTestImages(path):
    paths = imagePaths(path)
    images = openImages(paths)
    return images


trainPath = "Enter path to training images here"
testPath  = "Enter path to testing images here"

x,y = processTrainImages(trainPath)
x_train = x[:20000]
y_train = y[:20000]
x_dev = x[-5000:]
y_dev = y[-5000:]

x_test = processTestImages(testPath)
```

Displaying the shape

```python
np.shape(x_test)
```

# Making model


```python
layers = []
layers.append(keras.layers.Conv2D(6,3,padding = 'same',activation=tf.nn.relu))
layers.append(keras.layers.MaxPool2D())
layers.append(keras.layers.Conv2D(12,3,padding = 'same',activation=tf.nn.relu))
layers.append(keras.layers.MaxPool2D())
layers.append(keras.layers.Conv2D(24,3,padding = 'same',activation=tf.nn.relu))
layers.append(keras.layers.MaxPool2D())
layers.append(keras.layers.Conv2D(30,3,padding = 'same',activation=tf.nn.relu))
layers.append(keras.layers.MaxPool2D())
layers.append(keras.layers.Flatten())

for nodes in nodes_per_layer:
 layers.append(keras.layers.Dense(nodes, activation=tf.nn.relu)) # Using l2 regularization kernel_regularizer=tf.keras.regularizers.l2(l=0.0001)
 layers.append(keras.layers.Dropout(dropout_rate))
layers.append(keras.layers.Dense(2,activation='softmax'))
model = keras.Sequential(layers)
```

Compiling the model

```python
model.compile(optimizer=tf.train.AdamOptimizer(0.001), 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

# Training the Model

```python
hist = model.fit(x_train, y_train,epochs=10,batch_size=64,validation_data=(x_dev,y_dev))
```

# Printing the dev set Accuracy

```python
print('Finding dev accuracy..')
dev_loss, dev_acc = model.evaluate(x_dev, y_dev)
print('Dev accuracy:', dev_acc)
```

# Generating Test Set Predictions: 
##The test set is not annotated in this case

```python
test_predictions = model.predict(x_test)
print('Test predictions:', test_predictions)
```
